```python
import requests as re
```

## Sanction List, Program Names

```python
list_names = re.get("https://sanctionslistservice.ofac.treas.gov/sanctions-lists")
program_names = re.get("https://sanctionslistservice.ofac.treas.gov/sanctions-programs")
print(list_names.text)
print(program_names.text)
```

## Query from given list, or program

```python
def query_list(list_selection):
    
    list_return = re.get(f"https://sanctionslistservice.ofac.treas.gov/entities?list={list_selection}")
    print(len(list_return.text))
    
    return list_return.text

test_list = query_list("DPRK2")
```

```python
def query_program(prog_selection):
    
    prog_return = re.get(f"https://sanctionslistservice.ofac.treas.gov/entities?program={prog_selection}")
    print(len(prog_return.text))
    
    return prog_return.text

test_prog = query_program("DPRK2")
print(test_prog)
```

## Check Entity Count

```python
from bs4 import BeautifulSoup


def entity_count(xml_data): 
    
    soup = BeautifulSoup(xml_data, features = 'xml')

    matches = []
    
    # Define the regex pattern
    pattern = r'entity id="([^"]*)"'

    # Function to recursively search through all elements and attributes
    def search_element(element):
        
        # Check element attributes for the pattern
        for attr_value in element.attrib.values():
            match = pattern.search(attr_value)
            if match:
                matches.append(match.group(0))
        
        # Recursively search through child elements
        for child in element:
            search_element(child)

    # Start searching from the root element
    search_element(soup["sanctionsData"]["entities"])

    return matches


entities = entity_count(test_list)
print(len(entities))
```
